#Funções Built-in

abs() -> retorna o valor absoluto do numero

```python
abs(-9)
```

min() -> retorna o valor minimo de uma sequencia

```python
menorPreco = min(6,7,4,8,9,10)
menorPreco
```

max() -> retorna o valor máximo de uma sequencia

```python
maxPreco = max(6,7,4,8,9,10)
maxPreco
```

sum() -> faz a soma dos valores de uma sequencia

```python
somaTotal = sum([1,2,3])
somaTotal
```

pow() -> exponenciação

```python
2**3
```

```python
pow(2,3)
```

print() -> mostra um valor na saída

```python
print("esse é um texto")
```

```python
print(somaTotal)
```

input() -> lê um valor digitado no teclado pelo usuário

```python
nome = input("Qual o seu nome?")
```

```python
print(nome)
```

help() -> mostra as funcionalidades de uma determinada função

```python
help(max)
```
